1-carga de librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay
```

carga de datos

```python
data = load_breast_cancer()
X = data.data
y = data.target

print(f"Características: {data.feature_names}")
print(f"Objetivo: {data.target_names}")
```

4. Preprocesamiento de Datos

```python
train_data, test_data, train_labels, test_labels = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print(f"Tamaño del set de entrenamiento: {train_data.shape[0]} muestras")
print(f"Tamaño del set de prueba: {test_data.shape[0]} muestras")
```

5. Selección y Entrenamiento del Modelo

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(train_data, train_labels)
```

6. Prueba del Modelo y Resultados

```python
predictions = model.predict(test_data)

print("--- Reporte de Clasificación ---")
print(classification_report(test_labels, predictions, target_names=data.target_names))

print(f"Precisión (Accuracy): {accuracy_score(test_labels, predictions):.4f}")

cm = confusion_matrix(test_labels, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión")
plt.show()
```

7. Visualización del Árbol de Decisión

```python
plt.figure(figsize=(20,10), dpi=300)
plot_tree(model,
          feature_names=data.feature_names,
          class_names=data.target_names,
          filled=True,
          rounded=True)
plt.title("Visualización del Árbol de Decisión")
plt.show()
```
